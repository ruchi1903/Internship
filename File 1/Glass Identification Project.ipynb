```python
from collections import Counter

def outlier_hunt(data):
    outlier_indices = []
    for col in data.columns.tolist():
        Q1 = np.percentile(data[col], 25)
        Q3 = np.percentile(data[col], 75)
        IQR = Q3 - Q1
        outlier_step = 1.5 * IQR
        outlier_list_col = data[(data[col] < Q1 - outlier_step) | (data[col] > Q3 + outlier_step)].index
        outlier_indices.extend(outlier_list_col)
    
    outlier_indices = Counter(outlier_indices)
    multiple_outliers = [k for k, v in outlier_indices.items() if v > 2]
    return multiple_outliers

outliers = outlier_hunt(data[features])
print('The dataset contains %d observations with more than 2 outliers' % len(outliers))
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0,stratify=y)
y_train = y_train.values.ravel()
y_test = y_test.values.ravel()
print('Shape of X_train = ' + str(X_train.shape))
print('Shape of X_test = ' + str(X_test.shape))
print('Shape of y_train = ' + str(y_train.shape))
print('Shape of y_test = ' + str(y_test.shape))
```

```python
from sklearn.neighbors import KNeighborsClassifier

Scores = []

for i in range(2, 11):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    
    print(f"Training score for {i} neighbors: {knn.score(X_train, y_train)}")
    
    score = knn.score(X_test, y_test)
    Scores.append(score)

print("List of scores:", Scores)
```

```python
from sklearn.tree import DecisionTreeClassifier

Scores = []

for i in range(2, 11):
    tree = DecisionTreeClassifier(random_state=0)
    tree.fit(X_train, y_train)

    print(f"Training score for {i} iterations: {tree.score(X_train, y_train)}")

    score = tree.score(X_test, y_test)
    Scores.append(score)

print("List of scores:", Scores)
```

```python
from sklearn.linear_model import LogisticRegression

Scores = []

for i in range(2, 11):
    logistic = LogisticRegression(random_state=0, solver='lbfgs', multi_class='multinomial', max_iter=i)
    logistic.fit(X_train, y_train)

    print(f"Training score for max_iter={i}: {logistic.score(X_train, y_train)}")

    score = logistic.score(X_test, y_test)
    Scores.append(score)

print("List of scores:", Scores)
```

```python
from sklearn.svm import SVC

Scores = []

for i in range(1, 11):
    svc = SVC(C=i, gamma='auto')
    svc.fit(X_train, y_train)

    print(f"Training score for C={i}: {svc.score(X_train, y_train)}")

    score = svc.score(X_test, y_test)
    Scores.append(score)

print("List of scores:", Scores)
```
