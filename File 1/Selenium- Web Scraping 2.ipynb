#### Q1. Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.

```python
#Opening the shine page on automated chrome browser
driver.get("https://www.shine.com/")
```

```python
#Entering job title & location
job_title = driver.find_element(By.CLASS_NAME,"form-control ")
job_title.send_keys('Data Analyst')
```

```python
location = driver.find_element(By.XPATH, "/html/body/div/div[4]/div/div[2]/div[2]/div/form/div/div[1]/ul/li[2]/div/input")
location.send_keys('Bangalore')
```

```python
search = driver.find_element(By.CLASS_NAME,"searchForm_btnWrap_advance__VYBHN")
search.click()
```

```python
#scraping company name from page
company1 = driver.find_elements(By.XPATH, '//div[@class="jobCard_jobCard_cName__mYnow"]')
company1
```

```python
comp1 = []
for i in company1:
    comp1.append(i.text)
comp1
```

```python
#scraping job title from page
title1 = driver.find_elements(By.XPATH, '//h2[@itemprop="name"]')
title1
```

```python
jobtitle1 = []
for k in title1:
    jobtitle1.append(k.text)
jobtitle1
```

```python
#scraping job location from page
loc1 = driver.find_elements(By.XPATH, '//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2"]')
loc1
```

```python
Loc1 = []
for j in loc1:
    Loc1.append(j.text)
Loc1
```

```python
#scraping job experience from page
exp1 = driver.find_elements(By.XPATH, '//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t"]')
exp1
```

```python
Exp1 = []
for j in exp1:
    Exp1.append(j.text)
Exp1
```

```python
jobs1 = pd.DataFrame({})
jobs1['Job Title'] = jobtitle1[:10]
jobs1['Company Name'] = comp1[:10]
jobs1['Location'] = Loc1[:10]
jobs1['Job experience'] = Exp1[:10]
jobs1
```

#### Q8. Write a python program to scrape data for Top 1000 Quotes of All Time.

```python
search =  driver.find_element(By.XPATH, '//a[@href="/top_quotes.html"]')
search.click()
```

```python
quotes = driver.find_elements(By.XPATH, "//div[@class='wrap-blockquote']")
for quote in quotes:
    quote_text = quote.find_element(By.XPATH, ".//a[@class='title js-title']").text
    author = quote.find_element(By.XPATH, ".//a[@class='author js-author']").text
    type_of_quote = quote.find_element(By.XPATH, ".//a[@class='kw js-kw']").text
    print(f"Quote: {quote_text}\nAuthor: {author}\nType of Quote: {type_of_quote}\n")
```

#### Q9. Write a python program to display list of respected former Prime Ministers of India(i.e. Name, Born-Dead, Term of office, Remarks) from jagranjosh.com/.

```python
driver.get("https://www.jagranjosh.com/")
```

```python
gk_option = driver.find_element(By.XPATH, "//a[@title='General Knowledge']")
gk_option.click()
```

```python
pm_list = driver.find_element(By.XPATH, "//a[@title='List of all Prime Ministers of India']")
pm_list.click()
```

```python
pm_data = []
pm_table = driver.find_element(By.XPATH, "//table[@class='table table-bordered table-striped']")
pm_rows = pm_table.find_elements(By.XPATH, ".//tr")
for row in pm_rows:
    pm_cols = row.find_elements(By.XPATH, ".//td")
    if len(pm_cols) > 0:
        name = pm_cols[0].text
        born_dead = pm_cols[1].text
        term_of_office = pm_cols[2].text
        remarks = pm_cols[3].text
        pm_data.append([name, born_dead, term_of_office, remarks])
```

```python
pm_df = pd.DataFrame(pm_data, columns=["Name", "Born-Dead", "Term of Office", "Remarks"])
print(pm_df)
```

#### Q10: Write a python program to display list of 50 Most expensive cars in the world (i.e. Car name and Price) from https://www.motor1.com/

```python
driver.get("https://www.motor1.com/")
```

```python
search_bar = driver.find_element(By.XPATH, "//input[@class='search__input']")
search_bar.send_keys("50 most expensive cars")
```

```python
search_button = driver.find_element(By.XPATH, "//button[@class='search__button']")
search_button.click()
```

```python
car_data = []
car_table = driver.find_element(By.XPATH, "//table[@class='table table-bordered table-striped']")
car_rows = car_table.find_elements(By.XPATH, ".//tr")
for row in car_rows:
    car_cols = row.find_elements(By.XPATH, ".//td")
    if len(car_cols) > 0:
        car_name = car_cols[0].text
        car_price = car_cols[1].text
        car_data.append([car_name, car_price])
```

```python
car_df = pd.DataFrame(car_data, columns=["Car Name", "Price"])
print(car_df)
```
